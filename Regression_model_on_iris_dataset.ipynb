## Importing dependencies

```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn import metrics # for accuracy and precisio
```

## Data Collection and Preprocessing

```python
# loading data into pandas data frame

data = pd.read_csv("C://Users//DITU//Desktop//iris.csv")
data.head(10)
```

```python
# dataset shape 
data.shape
```

```python
# information about dataset
data.info()
```

```python
# summary of dataset 
data.describe()
```

```python
# Checking for null values

data.isnull().sum()
```

```python
sns.pairplot(data, hue = 'variety', vars = ['sepal.length','sepal.width', 'petal.length','petal.width'])
```

```python
x=data.iloc[:,:-4]
x.head()
```

```python
y=data['petal.length']
y.head()
```

## Visualize dataset 

```python
plt.figure(figsize=(15, 9))

plt.scatter(x,y, color= "red")
plt.xlabel("Sepal Length", fontsize = 20)
plt.ylabel("Petal Length", fontsize = 20)
plt.title(" Plot for Sepal Length and Petal Length ", fontsize = 25)
plt.show()
```

## Splitting the data into train and test dataset

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

## Model Training

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

## Model Evaluation

```python
training_data_prediction = model.predict(X_train)
```

```python
# R squared Error for Training data

training_error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared error : ", training_error_score)
```

```python
## Test Data Prediction
testing_data_prediction = model.predict(X_test)
```

```python
# R squared Error for Testing data

testing_error_score = metrics.r2_score(Y_test, testing_data_prediction)
print("R squared error : ", testing_error_score)
```

### Plot between training and testing data prediction

```python
plt.figure(figsize=(15, 9))

plt.scatter(X_train,Y_train, color= "green")
plt.xlabel("X train", fontsize = 10)
plt.ylabel("Y train", fontsize = 10)
plt.title(" Plot for x_train and y_train ", fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(15, 9))

plt.scatter(X_test,Y_test, color= "green")
plt.xlabel("X test", fontsize = 10)
plt.ylabel("Y test", fontsize = 10)
plt.title(" Plot for x_test VS y_test ", fontsize = 20)
plt.show()
```
